### Tab 2 - student demographics 
year, gender, study field

```python
# year, gender, study field
ds = df.groupby(by=['year', 'gender', 'study field']).count()[['ID']].rename(columns = {'ID':'count'})
ds
```

```python
dss = ds.reset_index().pivot_table(index=['year','gender'], 
                                   columns='study field', values='count').fillna(0).astype('int')
dss
```

```python
dss['all'] = dss.sum(axis=1)
dss.loc[('','total'),:] = dss.sum(axis=0)
dss = dss.astype('int')
dss
```

```python
## Latex code

subs = {'toprule':'hline', 'midrule':'hline', 'bottomrule':'hline'}

latex_code = dss.to_latex()
latex_code = replace_all(latex_code, subs)

print(latex_code)
```

### Tab 3 - distribution of achievers

```python
# year, study field, achiever
dach = df.groupby(by=['year','study field', 'achiever']).count()[['ID']].rename(columns = {'ID':'count'})
dach
```

```python
dachs = dach.reset_index().pivot_table(index=['year'], 
                                   columns=['study field','achiever'], values='count').fillna(0).astype('int')

dachs.loc['total',:] = dachs.sum(axis=0)
dachs['all'] = dachs.sum(axis=1)
dachs = dachs.astype('int')
dachs
```

```python
## Latex code

subs = {'toprule':'hline', 'midrule':'hline', 'bottomrule':'hline'}

latex_code = dachs.to_latex()
latex_code = replace_all(latex_code, subs)

print(latex_code)
```
