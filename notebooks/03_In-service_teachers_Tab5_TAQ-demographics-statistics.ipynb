# Data - wrangling

```python
file = '../data/03_In-service_teachers_attitudes-codes.xlsx'
filename = file.split('.')[0]
df = pd.read_excel(file, header=0)
show_allrowscols(df)
```

```python
n = df.shape[0]
n
```

```python
c = df.columns.tolist()
frameshow(c, transpose=True)
```

```python
dff = df.drop([c[2], c[7], c[21], *c[11:22], *c[24:]], axis=1)
dff
```

```python
c = dff.columns.tolist()
frameshow(c,transpose=True)
```

```python
# categorical variables
labels = ['strongly agree', 'agree', 'neutral', 'disagree', 'strongly disagree']
dff['next app extension'] = pd.Categorical(dff['next app extension'], categories=labels)


subs = {'certainly yes':'strongly agree', 'yes':'agree', "don't know":'neutral', 'no':'disagree', 'certainly no':'strongly disagree'}
dff[['future app use']] = dff[['future app use']].replace(subs)
dff['future app use'] = pd.Categorical(dff['future app use'], categories=labels)

labels = ['western','central', 'eastern']
col = 'region'
dff[col] = pd.Categorical(dff[col], categories=labels)

labels = ['less than 5',  '6-10', '11-15', '16-20', '21-30','31-40']
col = 'practice'
dff[col] = pd.Categorical(dff[col], categories=labels)

col = 'gender'
dff[col] = pd.Categorical(dff[col], categories=['F','M'])

dff
```

```python
dff.info()
```

# Summary - demographics

## gender vs. practice

```python
ds = dff.groupby(by=['practice', 'gender']).count()[[c[0]]].rename(columns = {c[0]:'count'})
ds.replace({np.nan:0}).astype('int')
```

```python
dss = ds.reset_index().pivot_table(index=['practice'], 
                                   columns='gender', values='count').fillna(0).astype('int')
```

```python
dss
```

```python
%matplotlib inline
#ticks = [' strongly \n agree','agree', 'neutral', 'disagree', '  strongly \n disagree']
g = dss.plot.bar(rot=0);
#g.set_xticklabels(ticks);
g.set_yticks([0,5,10,15,20]);
```

```python
%matplotlib inline
#ticks = [' strongly \n agree','agree', 'neutral', 'disagree', '  strongly \n disagree']
g = dss.plot.bar(rot=0, stacked=True);
#g.set_xticklabels(ticks);
g.set_yticks([0,5,10,15,20]);
```

```python
dss.index = dss.index.add_categories('total')
dss.columns = dss.columns.add_categories('all')
```

```python
dss.loc['total',:] = dss.sum()
```

```python
dss
```

```python
dss['all'] = dss.sum(axis=1)
dss = dss.astype('int')
dss
```

```python
subs = {'toprule':'hline', 'midrule':'hline', 'bottomrule':'hline'}

latex_code = dss.to_latex()
latex_code = replace_all(latex_code, subs)

print(latex_code)
```
