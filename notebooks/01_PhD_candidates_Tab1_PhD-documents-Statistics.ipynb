# Data

PhD project and thesis $-$ All PhD students who learned at least one Data Science tool

```python
# load data as dataframe df
file = '../data/01_PhD-documents.xlsx'
df = pd.read_excel(file)
df
```

```python
cols = list(df.columns)
print(cols)
```

```python
# rename columns - lower case, '_' instead white spaces
df.columns = np_clean_str(cols, whitespace='_', case='lower')
df.head()
```

```python
# set multindex 
multidx = ['nick', 'gender', 'phd', 'date']
df = df.set_index(multidx).sort_index()
df
```

```python
c = df.columns
frameshow(list(c), transpose=True)
```

```python
# groups of columns: all sw, spreadsheets, special sw, cloud, statistics
soft =c[:-2]
spread = [c[0],c[3]]
spsoft = [*c[1:3], *c[4:8]]
cloud = [c[3],*c[6:8]]
stat = c[-2:]
```

# Stat methods summary

```python
# sum of stat cols and its interpretation
df['stats']=df[stat].sum(axis=1).replace({0:'none', 1:'basic', 2:'adv.'})
df.drop(columns=stat, inplace=True)
```

```python
df
```

# Software summary

## Cloud

```python
# sum of cloud cols and its interpretation
df['cloud?'] = df.loc[:,cloud].sum(axis=1).apply(lambda x: 'local' if x==0 else 'cloud')
df[['cloud?']]
```

## Spreadsheets, special software

```python
# replace 1 by names of sw, 0 by nan
df.loc[:, soft] = df.loc[:, soft].replace(1, pd.Series(soft,soft)).replace({0:np.nan})
```

```python
# spreadsheets summary
combine = lambda row: ' '.join(row.dropna())
df['spreadsheets'] = df.loc[:,spread].apply(combine, axis=1).replace({'':'none'})

df.drop(columns=spread,inplace=True)

df[['spreadsheets']]
```

```python
# special software summary
df['special sw'] = df.loc[:, spsoft].apply(combine, axis=1).replace({'':'none'})
df.drop(columns=spsoft, inplace=True)
df[['special sw']]
```

# Tab 1

## Tab 1 - Final sw summary

```python
# rename using real names of sw
subs = ['excel', 'Excel', 'google_sheets', 'Google sheets']
df.spreadsheets = np_clean_str(df.spreadsheets, whitespace=',', replace = subs)

subs = ['rsrp', 'RSRP', 'xlstat', 'XLSTAT', 'spss', 'SPSS', 
        'physport', 'PhysPort', 'jupyter', 'Jupyter', 'r,', 'R,']
df['special sw'] = np_clean_str(df['special sw'], whitespace=',', replace = subs)

df = df.replace({'r':'R'})
df
```

```python
# removing projects if a student has PhD thesis
dff = df.reset_index().drop_duplicates(subset = multidx[:-2], keep='last').set_index('nick')
dff = dff.sort_values(by='date')
dff
```

```python
# adding phase columns - first 6 in phase I, remaining 3 in phase III
dff['phase'] = 6*['I']+3*['II']
dff = dff.reset_index().set_index(['phase','nick'])
dff
```
