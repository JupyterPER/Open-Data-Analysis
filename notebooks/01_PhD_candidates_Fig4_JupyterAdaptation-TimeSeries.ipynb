### Graphical parameters

```python
%matplotlib inline

import matplotlib.pyplot as plt
import matplotlib as mpl
#mpl.style.use('ggplot')

size = (10,6)
lcolor = ['lightgray','dimgray','dimgray','dimgray','lightgray']
sizebox = (6,6)
trans = 0.7
font ='large'
line = 0
resolution = 300
xlim = [pd.Timestamp('2017-05-01'), pd.Timestamp('2020-07-10')]
ext = '_short.png'
style = ['solid', 'dotted', 'dotted', 'dotted', 'solid']
#path = '/media/sf_D_DRIVE/Dropbox/00 publikacie/Didaktika/2019Jupyter/clanok/figures/'
```

### Location of Jupyter software

```python
# summarize total counts
cols = ['location']
df.groupby(by=cols).count()[['file']]
```

```python
ddf = df

# summarize monthly
dfstat = ddf.groupby(by=cols).resample('M').count()[['file']].reset_index()

# remove zero values, shorten date to month, drop date column
dfstat = dfstat[dfstat.file!=0]
dfstat = dfstat.set_index(dfstat.date.dt.to_period('M'))
dfstat = dfstat.drop(columns = 'date')

show_allrowscols(dfstat.transpose())
```

```python
# reshape table by pivoting 
dfg = dfstat.pivot_table(index='date', columns=cols, values='file').fillna(0)
show_allrowscols(dfg.transpose())
```

```python
# set average daily number of files
dfm = dfg[dfg.index>=startCoCalc].apply(lambda x: x/x.index.days_in_month)
```

### Location - time series plots

*Pandas stacked **area** plot* displays the evolution of the value of several groups on the same graphic. The values of each group are displayed on top of each other. It allows to check on the same figure the evolution of both the total of a numeric variable, and the importance of each group.

```python
# pandas stacked area plot
g = dfm.plot.area(figsize=size, alpha=trans, fontsize=font, linewidth=line)

# draw important events as vertical lines
for i, date in enumerate(startdates):
    g.axvline(date, color=lcolor[i], linestyle=style[i])

# set labels
g.set_xlabel("") 
g.set_ylabel("daily average number of files", fontsize=font)
g.set_xlim(*xlim)
g.legend(title ='cloud vs. local', fontsize=font, title_fontsize=font)
plt.tight_layout();

# save figure
#name = 'TS_cloud_local'+ext
#g.get_figure().savefig(name, dpi=resolution)
#g.get_figure().savefig(path+name, dpi=resolution)
```

```python
# boxplot summary
g = dfm.plot.box(figsize=sizebox, fontsize=font);

# set lables 
g.set_xlabel("cloud vs. local", fontsize=font)
g.set_ylabel("daily average number of files", fontsize=font);
```
