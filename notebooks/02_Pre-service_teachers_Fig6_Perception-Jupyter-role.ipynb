## Graphical parameters

```python
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import product

name = 'Viewpoint1.png'
typ = 'bar'
resolution = 300
font ='large'
size = (11.5,4.5)
legend_title = "Pre-service teachers' personal reflections"
path = '/media/sf_D_DRIVE/Dropbox/00 publikacie/Didaktika/2019Jupyter/clanok/figures/'
ticks = [' strongly \n agree','agree', 'neutral', 'disagree', '  strongly \n disagree']
fontsize = [10,11]
xoff, yoff = - 0.25, 0.3
colors  = ['tab:blue','tab:orange','tab:green']
n = 28
```

## Fig6 - perception of Jupyter by years

```python
fig, axes = plt. subplots(ncols=3, figsize=size, sharey='row')

for pos, col in enumerate(c[-3:]):
    dgf = dff.groupby(by=[col,'year']).count().iloc[:,0].unstack()
    gsum = dgf.sum(axis=1)
    g = dgf.plot.bar(ax=axes[pos], stacked=True, legend = False, 
                     fontsize=10, rot=0, width=0.7, color=colors, zorder=5)
    g.grid(axis='y')
    g.set_title(col)
    g.set_xticklabels(ticks);
    g.set_yticks(np.arange(0, 20, 2));
    g.set_ylim(top=18)
    g.set_xlabel('')
    g.set_ylabel("number of responses")
        
    for i, x in enumerate(np.arange(len(ticks))):
        y = gsum.iloc[i]
        pery = round(y/n*100)
        #if y>0:
        text = str(pery)[:2]+'%'
        if pery<10: text = ' '+str(pery)[:1]+'%'
        g.annotate(text, (x+xoff, y+yoff), fontsize=11, zorder=10) 

legend = fig.legend(labels = ['year 2017','year 2018','year 2019' ], 
          bbox_to_anchor=(0.05,1,0.925,0.1), 
          loc="lower center", mode="expand", 
          borderaxespad=0, ncol=3,
          title = legend_title,
          fontsize =font,
          title_fontsize =font
         )

legend.get_frame().set_edgecolor('black')
legend.get_frame().set_facecolor('lightgrey')

#plt.minorticks_off()
plt.tight_layout()


#save figure
#fig.savefig(name, bbox_extra_artists=(legend,), bbox_inches='tight', dpi=resolution)
#fig.savefig(path+name, bbox_extra_artists=(legend,), bbox_inches='tight', dpi=resolution)
```
