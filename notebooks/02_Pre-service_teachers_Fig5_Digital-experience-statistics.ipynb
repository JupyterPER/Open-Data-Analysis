## Merged dataframe

```python
df = pd.merge(dd,dff)
#df = pd.merge(dd, dff, left_on=['ID','study field'], right_on=['personal_ID', 'personal_study field'], how='left')
df.columns = np_clean_str(df.columns, whitespace = '_')
c = df.columns.to_list()
df.shape
```

## Setting categorical variables

```python
col, labels = 'gender', ['F', 'M']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'study_field', ['Math-Phys', 'Phys-Bio', 'Phys-Comp', 'Phys-Chem', 'Phys-Geo']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'achiever', ['low', 'high']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'city', ['0-9999', '10000-19999','20000-49999', '50000-99999', '100000+']
df[col] = pd.Categorical(df[col], categories=labels)
```

```python
col, labels = 'own_device_notebook', ['Win10', 'macOS', 'unspecified']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'own_device_smartphone', ['android', 'iPhone', 'unspecified']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'own_device_smartphone', ['android', 'iPhone', 'unspecified']
df[col] = pd.Categorical(df[col], categories=labels)

col, labels = 'own_device_tablet', ['android', 'iPad', 'Win10']
df[col] = pd.Categorical(df[col], categories=labels)
```

# Fig 5 - graphical summary together

## Graphical parameters

```python
titles = {'HS-DT':'DT used in high school math',
         'home-DT':'own DT available at home',
         'course-DT':'own DT available for the course',
         'brand-DT':'own DT brand'}
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
from itertools import product

name = 'DigitalExperience.png'
typ = 'bar'
resolution = 300
font ='large'
size = (10,10)
legend_title = "Pre-service teachers' digital experience before the course"
path = '/media/sf_D_DRIVE/Dropbox/00 publikacie/Didaktika/2019Jupyter/clanok/figures/'
fontsize = [10,11]
xoff, yoff = [-0.1,-0.15,-0.1,-0.1], 0.3
paltab = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','darkgrey']
palbase = ['b','y','g','r','m','darkgrey']
colors = 3*[paltab]+[palbase]
n = 33
showleg = [False, False, False, True]
```

## Fig 5 - plots together

```python
# initial plot setup
fig, axes = plt.subplots(nrows=2, ncols=2,  figsize=size, sharey='row')
position = list(product([0,1], repeat=2))
graphs = ['home-DT', 'HS-DT', 'course-DT', 'brand-DT']

for i, graph in enumerate(graphs):
    g = dg[graph].plot.bar(ax=axes[position[i]], stacked=True, 
                           legend = showleg[i], fontsize=11, rot=0, color=colors[i], zorder=5)
    gsum = dg[graph].sum(axis=1)
    g.grid(axis='y')
    g.set_title(titles[graph])
    g.set_xticklabels(ticks[graph])
    g.set_yticks(np.arange(0, 35, 5));
    g.set_ylim(top=35)
    g.set_xlabel('')
    g.set_ylabel("number of responses",fontsize =11)
        
    for i, x in enumerate(np.arange(len(ticks[graph]))):
        y = gsum.iloc[i]
        pery = round(y/n*100)
        #if y>0:
        text = str(pery)[:2]+'%'
        if pery==100: text ='100%'
        if pery<10: text = ' '+str(pery)[:1]+'%'
        g.annotate(text, (x+xoff[i], y+yoff), fontsize=11, zorder=10) 

legend = fig.legend(labels = ['year 2017','year 2018','year 2019' ], 
          bbox_to_anchor=(0.058,1,0.904,0.1), 
          loc="lower center", mode="expand", 
          borderaxespad=0, ncol=3,
          title = legend_title,
          fontsize =font,
          title_fontsize =font         )

legend.get_frame().set_edgecolor('black')
legend.get_frame().set_facecolor('lightgrey')

#plt.minorticks_off()
plt.tight_layout()


#save figure
#fig.savefig(name, bbox_extra_artists=(legend,), bbox_inches='tight', dpi=resolution)
#fig.savefig(path+name, bbox_extra_artists=(legend,), bbox_inches='tight', dpi=resolution)
```
